```python
import pymongo
import json
```

```python
with open('restaurants.json','r') as r:
    data = json.load(r)

url = 'mongodb+srv://admin:admin@cluster0.xmr3d6s.mongodb.net/?retryWrites=true&w=majority'
# url = f'{connection_type}+{user}+':'+{password}+{adress}+{resto}'
myclient = pymongo.MongoClient(url)

mydb = myclient["Cluster0"] # INSTANCIAR BASE DE DATOS
restaurantes = mydb['restaurantes']
restaurantes.insert_many(data)
```

```python
# Mostrar todos los documentos de la coleccin restaurantes
list(restaurantes.find())
```

```python
#Mostrar los campos restaurant_id, nombre, distrito y cocina, pero excluya el campo _id para todos los documentos de la coleccin restaurantes

list(restaurantes.find({'borough':'Bronx'},{'_id':0,'restaurant_id':1,'name':1,'borough':1,'cuisine':1}))
```

```python
# Mostrar los primeros 5 restaurantes que se encuentran en el distrito Bronx
list(restaurantes.find({'borough':'Bronx'},{'_id':0}).limit(5))
```

```python
# Devolver los restaurantes que lograron una puntuacin superior a 80 pero inferior a 100
    
list(restaurantes.find({ 'grades': { '$elemMatch': {'score':{ '$gt': 80, '$lt': 100 } }}},
    {'_id':0,
      'grades': {
        '$filter': {
          'input': '$grades',
          'as': 'grade',
          'cond': {
            '$and': [
              { '$gt': ['$$grade.score', 80] },
              { '$lt': ['$$grade.score', 100] }
            ]
          }}}}
))
```

```python
# Devolver los restaurantes que se ubican en un valor de latitud inferior a -95.754168

list(restaurantes.find({ 'address.coord.0': { '$lt': -95.754168} },
{'_id':0,'name':1}
))
```

```python
# Devolver los restaurantes que no preparan cocina americana y lograron una puntuacin superior a 70 y se ubicaron en una longitud inferior a -65.754168. Nota: Realice esta consulta sin usar el operador $and

where = {
    'cuisine': {'$ne': 'American '},
    'grades.score': {'$gt': 70},
    'address.coord.0': {'$lt': -65.754168}
    
}

projection = {'_id': 0, 'name': 1}

list(restaurantes.find(where, projection))
```

```python
# Devolver los restaurantes que no preparan cocina americana y lograron un punto de calificacin 'A' que no pertenece al distrito de Brooklyn. El documento debe mostrarse segn la cocina en orden descendente.

where = {
    'cuisine': {'$ne': 'American '},
    'grades.grade': {'$eq': 'A'},
    'borough': {'$ne': 'Brooklyn'}
    
}

projection = {'_id': 0, 'borough':1,'name': 1}

list(restaurantes.find(where, projection).sort('cuisine', -1))
```

```python
# Devolver los restaurantes que pertenecen al distrito Bronx y preparan platos americanos o chinos

where = {
    'cuisine': {'$in': ['American ','Chinese']},
    'borough': {'$eq': 'Bronx'}
    
}

projection = {'_id': 0, 'borough':1,'cuisine':1,'name': 1}

list(restaurantes.find(where, projection))
```

```python
# Devolver ID del restaurante, nombre, distrito y la cocina para aquellos restaurantes que pertenecen al distrito de Staten Island o Queens o Bronx o Brooklyn


where = {

'borough': {'$in': ['Bronx', 'Staten Island', 'Queens', 'Brooklyn']}
    
}

projection = {'_id': 0,'restaurant_id': 1, 'borough':1,'cuisine':1,'name': 1}

(list(restaurantes.find(where, projection)))
```

```python
# Devolver ID del restaurante, nombre, distrito y la cocina de aquellos restaurantes que lograron una puntuacin que no supere los 10

where = {

'grades.score': {'$not': {'$gt':10}}
    
}

projection = {'_id': 0,'restaurant_id': 1, 'borough':1,'cuisine':1,'name': 1}

list(restaurantes.find(where, projection))
```

```python
# Devolver ID del restaurante, nombre, direccin y ubicacin geogrfica del restaurante de aquellos donde el segundo elemento de la matriz coord contiene un valor que es ms de 42 y hasta 52

where = {
    'address.coord.1': {
        '$gt': 42,
        '$lt': 53
    }
}

projection = {'_id': 0, 'address': 1, 'name': 1}

list(restaurantes.find(where, projection))
```
